```python
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator = np.std)# default estimator is mean
```

```python
sns.countplot(x='sex',data = tips)
```

```python
sns.countplot(x='time', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.displot(tips['total_bill'],bins = 30,kde=True)
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.pairplot(tips,hue='sex')
```

```python
tips_corr = tips.corr()
```

```python
print(tips_corr)
```

```python
sns.heatmap(tips_corr,annot=True)
```

```python
flight = sns.load_dataset('flights')
```

```python
flights_pivot = flight.pivot_table(index='month',columns='year',values='passengers')
```

```python
flights_pivot
```

```python
sns.heatmap(flights_pivot)
```
