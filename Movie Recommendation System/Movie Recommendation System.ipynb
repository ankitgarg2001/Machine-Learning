# Creating Movie Recommendation

```python
moviemat = df.pivot_table(index="user_id",columns="movie_name",values="rating")
```

```python
ratings.sort_values('user_id',ascending=False)
star_wars_user_rating = moviemat['Star Wars (1977)']
```

```python
similar_to_star_wars = moviemat.corrwith(star_wars_user_rating)
```

```python
similar_to_star_wars
```

```python
correlation_starwars = pd.DataFrame(similar_to_star_wars,columns = ['correlation'])
```

```python
correlation_starwars.dropna(inplace=True)
```

```python
correlation_starwars
```

```python
correlation_starwars.sort_values(by='correlation',ascending=False).head(10)
```

```python
correlation_starwars = correlation_starwars.join(ratings['user_id'])
```

```python
correlation_starwars = correlation_starwars[correlation_starwars['user_id']>100]
```

```python
correlation_starwars.sort_values('correlation',ascending=False).head(10)
```

```python
def getSimilarMovies(title):
    similar_movie = pd.DataFrame(moviemat.corrwith(moviemat[title]),columns=['Correlation'])
    similar_movie.dropna(inplace=True)
    similar_movie = similar_movie.join(ratings['user_id'])
    similar_movie = similar_movie[similar_movie['user_id']>100].sort_values('Correlation',ascending=False).head(10)
    return similar_movie
```

```python
getSimilarMovies('Star Wars (1977)')
```

```python
getSimilarMovies('Titanic (1997)')
```
