# <font color='green'>Data Visualisation</font>

```python
import numpy as np
import matplotlib.pyplot as plt
```

# <font color='green'>Line Plot</font>

```python
# Line Plot
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
themes = plt.style.available
print(themes)
```

```python
plt.style.use('seaborn')
plt.plot(x,y1,'r',label = "Apple",marker = 'o')
plt.plot(x,y2,'g',label = "Kiwi",linestyle = "dashed",marker = '*')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Price')
plt.title("Prices of Fruits over time")
plt.show()
```

```python
prices = np.array([1,2,3,4])**3
```

```python
plt.plot(prices)
plt.show()
```

# <font color='green'>Scatter Plots</font>

```python
plt.figure(figsize=(5,5))
plt.scatter(x,y1,color = 'r',label = "Apple",marker = '^')
plt.scatter(x,y2,color = 'g',label = "Kiwi",linestyle = "dashed")
plt.legend()
plt.xlabel('Time')
plt.ylabel('Price')
plt.title("Scatter Plot - Prices of Fruits over time")
plt.show()
```

# <font color = 'green'>Bar-graphs</font>

```python
plt.style.use("dark_background")
x_coordinates = np.array([0,1,2])*2
plt.bar(x_coordinates,[10,20,15],width=0.5,label='Current Year',tick_label=["Gold","Platinum","Silver"]) # Current year prices
plt.bar(x_coordinates+0.5,[20,10,12],width=0.5,label='Next Year') # Next year prices
plt.title("Metal Price Comparision")
plt.xlabel("Metal")
plt.ylabel("Price")
plt.ylim(0,40)
plt.xlim(-2,5)
plt.legend()
plt.show()
```

# <font color='green'>Pie Chart</font>

```python
subjects = "Maths","Chem","Physics","English"
weightage = [20,10,15,5]
plt.pie(weightage,labels=subjects,explode=(0,0,0.1,0),autopct='%1.1f%%')
plt.show()
```

# <font color='green'>Histogram</font>

```python
Xsn = np.random.randn(100)
sigma = 8
u = 70
X = np.round(Xsn*sigma+u)
X2 = np.round(Xsn*5 + 40)
print(X)
```

```python
plt.style.use('seaborn')
plt.hist(X,label = 'Marks in Maths')
plt.hist(X2,label = 'Marks in Physics',alpha=0.8)
plt.ylabel('Prob/Freq of students within that marks range')
plt.xlabel('Marks range')
plt.legend()
plt.title('Histogram')
plt.show()
```

# <font color='green'>Data Visualization Challenge</font>
Given a Movies Dataset, your tasks is to make a visualization that-
- Plots the Length of the Movie Title Names on X-axis
- Frequency Counts on Y-axis ie Number of Movies havimg 'x' characters in their title

```python
import pandas as pd
```

```python
df = pd.read_csv('movie-dataset.csv')
```

```python
 df
```

```python
titles = list(df['movie_title'])
```

```python
print(titles[:5])
```

```python
freq_title = {}
for title in titles:
    length = len(title)
    if freq_title.get(length) is None:
        freq_title[length] = 1
    else:
        freq_title[length] += 1
```

```python
print(freq_title)
```

```python
x = np.array(list(freq_title.keys()))
y = np.array(list(freq_title.values())) 
```

```python
plt.scatter(x,y)
plt.xlabel('Length of movie titles')
plt.ylabel('Freq of each length')
plt.show()
```
